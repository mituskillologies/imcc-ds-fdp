```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('datasets/mtcars.csv')
x = df[['disp','hp','wt']]
y = df['mpg']
reg = LinearRegression()
reg.fit(x, y)
print('Accuracy:', reg.score(x, y))
```

```python
x = df.drop('mpg',axis=1)
```

```python
from sklearn.feature_selection import RFE
rfe = RFE(estimator=reg, n_features_to_select=3)
rfe.fit(x,y)
```

```python
rfe.get_support()
```

```python
x = x.iloc[:,rfe.get_support()]
x
```

```python
reg.fit(x, y)
print('Accuracy:', reg.score(x, y))
```

```python
rfe.ranking_
```

#### chi2 method

```python
# Dataset: wisc_bc_data.csv
# Location: https://mitu.co.in/dataset
```

```python
df = pd.read_csv('datasets/wisc_bc_data.csv')
```

```python
df
```

```python
x = df.iloc[:,2:]
y = df['diagnosis']
```

```python
x.shape
```

```python
y.value_counts()
```

```python
from sklearn.feature_selection import chi2, SelectKBest 
```

```python
skf = SelectKBest(chi2, k=10)
```

```python
skf.fit(x, y)
```

```python
skf.get_support()
```

```python
x = x.iloc[:,skf.get_support()]
```

```python
x.shape
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    x, y, random_state=0, test_size=0.25)

clf = LogisticRegression()
clf.fit(x_train, y_train)
y_pred = clf.predict(x_test)
print('Accuracy:', accuracy_score(y_test, y_pred))
```
