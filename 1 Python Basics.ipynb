#### Operators

```python
# Arithmetic Operators
num1 = 67
num2 = 19
```

```python
print(num1+num2)
print(num1-num2)
print(num1*num2)
print(num1/num2)
print(num1//num2)  # floor division
print(num1%num2)
print(num1**num2)  # power of
```

```python
# Relational Operators
print(num1>num2)
print(num1>=num2)
print(num1<num2)
print(num1<=num2)
print(num1==num2)
print(num1!=num2)
```

```python
# Logical operators
print(num1>num2 and num1 < 100)
print(num1<num2 or num1 < 100)
print(not num1>num2)
```

```python
# Assignment operator
print(num1)
num1 += 2
print(num1)
```

#### decision making

```python
num = int(input('Enter the number: '))
if num > 0:    # Compulsory, Condition, Non-repeatable
    print('Positive Number')
elif num ==0:  # Optional, Condition, Repeatable
    print('It is Zero')
else:          # Optional, Non-condtion, Non-repeatable
    print('Negative Number')
```

###  Object, Class

```python
class Person:
    def __init__(self,n,a):  # Constructor
        self.name = n        # class variables / instance variable
        self.age = a
    def show(self):          # class method
        print('Values are:')
        print(self.name, self.age)
```

```python
obj1 = Person('Ajay',25)
```

```python
obj1.show()
```

```python
obj1.age = 27
obj1.show()
```
